```python
import pandas as pd
data = pd.read_csv("/content/superstore-dataset-final/Sample - Superstore.csv", encoding='latin1')
```

```python
data.head()
```

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files

sns.set_style("whitegrid")
```

```python
print(data.shape)

data.info()

data.describe()

data.isnull().sum()
```

```python
#duplicates record
duplicates = data.duplicated().sum()

print("Duplicate Rows:",duplicates)

data = data.drop_duplicates()
```

```python
#Missing values
data.isnull().sum()
#Fill Numerical
num_cols = data.select_dtypes(include=np.number).columns

for col in num_cols:
    data[col].fillna(data[col].median(), inplace=True)
#Fill categorical
cat_cols = data.select_dtypes(include="object").columns

for col in cat_cols:
    data[col].fillna(data[col].mode()[0], inplace=True)
#Check Again
data.isnull().sum()
```

```python
#Outlierdetection
plt.figure(figsize=(10,5))

sns.boxplot(data=data.select_dtypes(include=np.number))

plt.xticks(rotation=45)

plt.show()

#Remove outliers using IQR
numeric =  data.select_dtypes(include=np.number)

Q1 = numeric.quantile(.25)

Q3 = numeric.quantile(.75)

IQR = Q3-Q1

data = data[~((numeric<(Q1-1.5*IQR)) | (numeric>(Q3+1.5*IQR))).any(axis=1)]
```

```python
#Final dataset
data.shape
```

```python
import os

os.makedirs("cleaned_data",exist_ok=True)

data.to_csv("cleaned_data/cleaned_sales_data.csv",index=False)
```

```python
#Sales by category
plt.figure(figsize=(10,5))

sales = data.groupby("Category")["Sales"].sum()

sales.plot(kind="bar")

plt.title("Sales by Category")

plt.xlabel("Category")

plt.ylabel("Sales")

plt.tight_layout()

plt.savefig("sales_by_category.png")

plt.show()
```

```python
#Monthly Sales Trend
data["Order Date"] = pd.to_datetime(data["Order Date"])

monthly = data.groupby(data["Order Date"].dt.to_period("M"))["Sales"].sum()

monthly.index = monthly.index.astype(str)

plt.figure(figsize=(12,5))

monthly.plot()

plt.title("Monthly Sales")

plt.xlabel("Month")

plt.ylabel("Sales")

plt.savefig("monthly_sales.png")

plt.show()
```

```python
#Top 10 Products
top =  data.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))

top.plot(kind="barh")

plt.title("Top Selling Products")

plt.savefig("top_products.png")

plt.show()
```

```python
#Coorelation heatmap
plt.figure(figsize=(8,6))

sns.heatmap(data.select_dtypes(include=np.number).corr(),
            annot=True,
            cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.savefig("correlation_heatmap.png")

plt.show()
```

```python
#Dashboard
fig,axs=plt.subplots(2,2,figsize=(16,10))

# Sales by Category
sales.plot(kind="bar",ax=axs[0,0])
axs[0,0].set_title("Sales by Category")

# Monthly
monthly.plot(ax=axs[0,1])
axs[0,1].set_title("Monthly Sales")

# Top Products
top.plot(kind="barh",ax=axs[1,0])
axs[1,0].set_title("Top Products")

# Heatmap
sns.heatmap(data.select_dtypes(include=np.number).corr(),
            annot=True,
            cmap="coolwarm",
            ax=axs[1,1])

plt.tight_layout()

plt.savefig("dashboard.png")

plt.show()
```
